```python
import folium
import requests
import pandas

arrest_table = pandas.read_csv("https://cmsc320.github.io/files/BPD_Arrests.csv")

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)

arrest_table.dropna(subset = ['incidentLocation', 'race', 'sex'], inplace = True)
arrest_table
#arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
```

For my dataset, I chose Baltimore PD arrest data. In this dataset, we are shown information on the suspects (age, race, gender), time of arrest, location, type of crime, and other information that explains these categories further.

```python
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=13)
map_osm
```

```python
arrest_table.columns
```

```python
for index, arrest in arrest_table.iterrows():
  if index < 1000:
      if arrest["incidentOffense"] == "87-Narcotics":
          folium.Marker(location=[arrest["lat"], arrest["long"]],
                      icon=folium.Icon(color='red')).add_to(map_osm)
          
      if arrest["incidentOffense"] == '97-Search & Seizure':
          folium.Marker(location=[arrest["lat"], arrest["long"]],
              icon=folium.Icon(color='blue')).add_to(map_osm)
map_osm 
```

With this map, I show a portion of narcotic and search and seizure arrests. Because of computational power limitations, as well as improving the readability of the map by not having too many markers, I used a sample of the arrests (index < 1000). This map helps visualize the location of narcotic-related crimes, including search and seizure due to many such instances being seizure of narcotics or paraphernalia. Through the visualization of our data, we can conclude that narcotic-related crimes appear to be more frequent in West Baltimore (a la hit TV show "The Wire").
